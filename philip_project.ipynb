```python
# Research questions
#1. top 10 states with highest price variations
#2. top 10 countries with highest price variations
#3. Distribution of market types
#4. Price variation by currency
#5. Top 10 produce
```

```python
# Calculate price variation for each state
state_price_variations = df.groupby('state')['price'].std()

# Find top 10 states with highest price variations
top_states_variation = state_price_variations.nlargest(10)

plt.figure(figsize=(10, 6))
top_states_variation.plot(kind='bar')
plt.title('Top 10 States with Highest Price Variations')
plt.xlabel('State')
plt.ylabel('Price Variation (Standard Deviation)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Calculate price variation (standard deviation) for each country
country_price_variations = df.groupby('country')['price'].std()

# Find top 10 countries with highest price variations
top_countries_variation = country_price_variations.nlargest(10)

# Create a bar plot to visualize the top 10 price variations by country
plt.figure(figsize=(10, 6))
top_countries_variation.plot(kind='bar')
plt.title('Top 10 Price Variations by Country')
plt.xlabel('Country')
plt.ylabel('Price Variation (Standard Deviation)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Count the occurrences of each market type
market_type_counts = df['market_type'].value_counts()

# Create a pie chart to visualize the distribution of market types
plt.figure(figsize=(8, 8))
plt.pie(market_type_counts, labels=market_type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Market Types')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Create box plot for price variation by currency
plt.figure(figsize=(10, 6))
sns.boxplot(x='currency', y='price', data=df)
plt.title('Price Variation by Currency')
plt.xlabel('Currency')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Calculate the counts of each produce
produce_counts = df['produce'].value_counts()

# Find top 10 produces
top_produces = produce_counts.nlargest(10)

# Create a bar plot to visualize the top 10 produces
plt.figure(figsize=(12, 8))
top_produces.plot(kind='bar', color='skyblue')
plt.title('Top 10 Produces')
plt.xlabel('Produce')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
